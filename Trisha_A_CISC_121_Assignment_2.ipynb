*   Constraints: −1000 <=a, b <= 1000 ; Cannot use + and - operators
*   Output: Integer (sum of a and b)

Pseudocode:
* Start with two integer, a and b
* While b is not 0:
  * Find where both numbers ahve 1s in the same place (the carry)
  * Add the numbers wothout carrying using XOR
  * Move the carry ti where it should be added next
* When b becomes 0, stop
* Final value stored in a (this is the sum of both numbers)



## 2. Peer Review & Draft Code
Leetcode solution used: https://leetcode.com/problems/sum-of-two-integers/solutions/4587116/detailed-explanation-with-numerical-examples-for-python/

Through this explanation, I learned how XOR and AND can be used to perform integer addition without relying on the + or - operators. XOR (^) adds two bits but ignores any carry. AND (&), on the other hand, identifies where both bits are 1, which is where a carry would occur. By repeatedly shifting the carry and combining it with XOR results, we can simulate the addition process bit by bit. The << operator shifts all bits to the left, effectively moving the carry to the next higher place value. Using this knowledge, I was able to complete Part 1, “Understanding and Breakdown,” with more accuracy.

```python
# Your draft code here
def get_sum(a,b):
  if not (-1000 <= a <= 1000 and -1000 <= b <= 1000):
    return "Error: values must be between -1000 and 1000"
  while b!=0:
    carry = a & b <<1
    a = a^b
    b = carry
  return a

  print(get_sum(2,3))
  print(get_sum(5,6))
  print(get_sum(-10,19))
```

Pseudocode:
*  Set the minimum value (-2^31) and the maximum value (2^31-1) to the constraints

*   If no remaning digits, the result is the last digit
