#  Problem 5 - Sort Colors




## 1. Understanding & Breakdown
Note: I used the LeetCode solution (linked below in part 2) to guide my pseudocode. The explanation helped me clearly understand the Dutch National Flag partitioning method and how it is applied to sort the colours efficiently.

* Input: Array (nums) of length n
  * 0 = red
  * 1 = white
  * 2 = blue
* Output: Modified array, such that 0's come first, followed by 1s and then followed by 2s
* Constraints:
  * n == nums.length
  * 1 <= n <= 300
  * nums[i] is either 0, 1, or 2
  * do not use sort function

Pseudocode:
* Set 3 pointers
  * Start: beginning of array
  * Current: the current index being checked
  * End: end of array
* Loop while current is less than end
* If current value is 0: swap value with start; and move the start pointer forward
* If current value is 1: only move current forward
* If current value is 2: swap value with end; and move the end pointer backwards

## 2. Peer Review & Draft Code


I reviewed the following solution under the LeetCode ‚ÄúSolutions‚Äù tab: https://leetcode.com/problems/sort-colors/solutions/6754733/two-pointers-1-pass-using-dutch-partitio-w4m1/.

The explanation and the mapping showing how this method sorts colours provided me with a very clear understanding of how the Dutch Flag algorithm works. This method seperated the array into 3 regions (one for 0s, one for 1s and, one for 2s) and uses 3 pointers to swap elements into to the correct area. As the current pointer moves through the list, 0s are moved to the front, 2s are moved to the end and the 1s remain in the middle.

```python
#Draft Code
def sortcolours(nums):
  if not (1 <= len(nums) <= 300):
    raise ValueError("Array length must be between 1 and 300.")

  for x in nums:
    if x not in (0, 1, 2):
      raise ValueError("Array elements must be 0, 1, or 2.")

  start, current, end = 0,0, len(nums)-1

  while current < end:
    if nums[current] == 0:
      nums[start], nums [current] = nums [current], nums[start]
      start +=1

    elif nums[current] ==1:
      current +=1

    else:
      nums[current], nums[end] = nums[end], nums[current]
      end -=1

    return nums

print(sortcolours([2,0,2,1,1,0])) # [0, 0, 1, 1, 2, 2]
print(sortcolours([2,0,1])) # [0, 1, 2]
```
