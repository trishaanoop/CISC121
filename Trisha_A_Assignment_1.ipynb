I reviewed the following solution under the LeetCode "Solutions" tab: https://leetcode.com/problems/valid-palindrome/solutions/7197820/clean-simple-efficient-beginner-friendly-easy-explanation/

From this source, I learned about the isalnum() method. I further clarified its use through a Stack Overflow post: https://stackoverflow.com/questions/43158093/how-to-check-if-a-string-contains-only-lower-case-letters-and-numbers, , which helped me understand that isalnum() checks whether the characters in a string are alphanumeric.

```python
# Your draft code here (does not need to be fully correct)
def checkPalindrome(s):
  new=""
  for i in s:
    if i.isalnum():
      new.append(i)

  cleaned = "".join(new) + " "
  return cleaned == cleaned[::-1]
print(checkPalindrome("race a car"))
print(checkPalindrome("A man, a plan, a canal: Panama"))
```

## 3. Final Correct Code



```python
# Final working solution
# [AI Disclaimer]: ChatGPT was used with the prompt "how to convert the code into lowecase"
# [AI Disclaimer]: ChatGPT was also used with the prompt "what line of code is causing a false output"
# It identified the issue in the line 'cleaned = "".join(new)', where I had mistakenly added an extra space at the end. This redundant space caused the palindrome check to return an incorrect result.
# [Source Used]: Source below was used to learn that .append() only works with a list instead of a string, which is why the draft produced an error
# Source: https://leetcode.com/problems/valid-palindrome/solutions/7197820/clean-simple-efficient-beginner-friendly-easy-explanation/
def checkPalindrome(s):
  new=[]
  for i in s:
    if i.isalnum():
      new.append(i.lower())

  cleaned ="".join(new)
  return cleaned == cleaned[::-1]
print(checkPalindrome("race a car"))
print(checkPalindrome("A man, a plan, a canal: Panama"))
```
